```python
import pandas as pd
import numpy as np
import seaborn as sns
from matplotlib import pyplot as plt
```

```python
df = pd.read_csv(r"C:\Users\Andrey\Downloads\33890050.dat", sep="\t", encoding = 'cp1251', skiprows=29, parse_dates=[['Date', 'Time']])
df
```

```python


fig, axs = plt.subplots(3, 1, figsize=(20, 10))
names = ["1", "2", "3"]

comp = ["Z", "N", "E"]
for i in range(3):
    buff = df[names[i]]
    N = 351
    m = np.convolve(buff, np.ones(N)/N, mode='valid')
    m = buff[:len(m)]-m
    axs[i].plot(df["Date_Time"][:len(m)], m, 'k')
    axs[i].grid()
    axs[i].set_xlabel("UTC Datetime")
    axs[i].set_ylabel(comp[i]+" компонента, у.е. АЦП")
    axs[i].autoscale(enable=True, axis='y', tight=True)
    axs[i].autoscale(enable=True, axis='x', tight=True)
#     axs[i]
plt.show()
```

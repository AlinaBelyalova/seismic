```python
import pandas as pd
import numpy as np
from matplotlib import pyplot as plt

# %matplotlib ipympl
```

```python
path = r"\\server\ngen_3_ВЫПОЛНЕНИЕ\2023.09 - ТСМ, Этап 2 Земполотно, (МСП, ГИС)\040 - Обработка\Инклинометрия\ПК1900"
skv1 = pd.read_csv(path+"\\ГФ 10.13.txt", sep='\t')
skv2 = pd.read_csv(path+"\\ГФ 10.14.txt", sep='\t')

skv2["E"] -= skv1["E"][0]
skv2["N"] -= skv1["N"][0]
skv1["E"] -= skv1["E"][0]
skv1["N"] -= skv1["N"][0]
```

```python
f = plt.figure(figsize=(12, 12), dpi=300)
# ax.patch.set_alpha(0.1)
ax = f.add_subplot(projection='3d')

ax.xaxis.pane.set_edgecolor('#000000')
ax.yaxis.pane.set_edgecolor('#000000')
ax.zaxis.pane.set_edgecolor('#000000')
ax.xaxis.pane.set_alpha(0.1)
ax.yaxis.pane.set_alpha(0.1)
ax.zaxis.pane.set_alpha(0.1)


# Now set color to white (or whatever is "invisible")
ax.xaxis.pane.set_edgecolor('k')
ax.yaxis.pane.set_edgecolor('k')
ax.zaxis.pane.set_edgecolor('k')




ax.plot(skv1["E"], skv1["N"], skv1["Z"])
ax.plot(skv2["E"], skv2["N"], skv2["Z"])
ax.scatter(skv1["E"], skv1["N"], skv1["Z"])
ax.scatter(skv2["E"], skv2["N"], skv2["Z"])
ax.scatter(skv1["E"][0], skv1["N"][0], skv1["Z"][0], s=150, facecolors='k', edgecolors='k', marker='v')
ax.scatter(skv2["E"][0], skv2["N"][0], skv2["Z"][0], s=150, facecolors='k', edgecolors='k', marker='v')
ax.text(skv1["E"][0]+0.5, skv1["N"][0], skv1["Z"][0], "ГФ 10.13", fontdict={"fontsize" : 12})
ax.text(skv2["E"][0]+0.5, skv2["N"][0], skv2["Z"][0], "ГФ 10.14", fontdict={"fontsize" : 12})
for i in range(0, skv1.values.shape[0], 1):
#     for j in range(0, skv2.values.shape[0], 5):
    ax.plot([skv1["E"][i], skv2["E"][i]] , [skv1["N"][i], skv2["N"][i]], [skv1["Z"][i], skv2["Z"][i]], c=
           'k', alpha=0.3)

ax.patch.set_edgecolor('black')
ax.set_aspect('equal')
ax.view_init(elev=35, azim=135+90, roll=0)

# ax = f.add_subplot(ncols=2)
plt.tight_layout()
# plt.show()
```

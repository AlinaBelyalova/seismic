```python
import segyio
import numpy as np
import matplotlib.pyplot as plt
```

```python
filename = '61_80.sgy'
with segyio.open(filename, ignore_geometry=True) as segyfile:
    segyfile.mmap()
    
    source_Elev = segyfile.attributes(segyio.TraceField.SourceSurfaceElevation)[:] #     
    rec_Elev = segyfile.attributes(segyio.TraceField.ReceiverGroupElevation )[:] #     
    sc = segyfile.attributes(segyio.TraceField.SourceGroupScalar )[:] #   
    if sc[0] < 0:
        sc = 1./np.abs(sc)
        
    su = np.unique(source_Elev)
    source_Elev = source_Elev*sc
    rec_Elev = rec_Elev*sc
#     print(np.unique(rec_Elev))
    buff_sou = source_Elev[rec_Elev == -67.55]
    buff_rec = rec_Elev[rec_Elev == -67.55]
#     print(buff_sou)
    
    samples= segyfile.trace.raw[:][rec_Elev == -67.55, :]
    samples = samples.T
    clip = 0.5
    for i in range(0, samples.shape[1]):
        samples[:,i] = samples[:,i] / max(np.abs(samples[:,i]))
#     plt.imshow(samples, cmap=plt.cm.seismic, vmin = - clip, vmax = clip)
    
    pick = segyfile.attributes(segyio.TraceField.ShotPoint)[:] #       ShotPoint 
    pick_buff = pick[rec_Elev == -67.55]
    pick_buff = pick_buff/100
    pick_buff[pick_buff<0] = np.nan
    
    
    dt = segyio.tools.dt(segyfile)/1000
    taxis = np.arange(0,samples.shape[0]*dt, dt)
    xaxis = buff_sou
    
    plt.ylim(0,100)
    plt.xlim(buff_sou[0], buff_sou[-1])
    plt.pcolormesh(xaxis,taxis,samples, cmap='gray', vmin=-clip, vmax=clip)
    plt.plot(buff_sou, pick_buff, color='red', linewidth=1.)
    plt.xlabel(' , ')
    plt.ylabel(', , ')
    plt.gca().invert_yaxis()
   
```

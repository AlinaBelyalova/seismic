```python
import pandas as pd
import numpy as np
import seaborn as sns
%matplotlib ipympl
from matplotlib import pyplot as plt
import matplotlib.dates as mdates
import datetime
import os
```

```python
file_paths = [
    r"C:\Users\Alina\Desktop\3388\33880380.dat",
    r"C:\Users\Alina\Desktop\3389\33890382.dat",
    r"C:\Users\Alina\Desktop\3390\33900381.dat",
    r"C:\Users\Alina\Desktop\3392\33920382.dat"
]

dfs = [pd.read_csv(file, sep="\t", encoding='cp1251', skiprows=29, parse_dates=[['Date', 'Time']]) for file in file_paths]
```

```python
names = ["3"]
colors = ['b', 'k', 'r', 'g']
comp = ["Z", "N", "E"]


plt.figure(figsize=(10, 4))

for i, df in enumerate(dfs):
    
    buff = df[names[0]]
    
    N = 50001
    if i == 0:
        
        buff = np.roll(buff, 50)
        
        
    m = np.convolve(buff, np.ones(N)/N, mode='valid')
    m = buff[:len(m)] - m
    
#     m = buff
#     m = m / np.max(np.abs(m))

    combined_datetime = pd.to_datetime(df['Date_Time']) + pd.to_timedelta(df['msec'], unit='ms')
    parent_dir = os.path.basename(os.path.dirname(file_paths[i]))

    plt.plot(combined_datetime[:len(m)], m, color=colors[i], label=parent_dir, linewidth=0.75, alpha=1)

    start_time = "2024-02-02 10:17:27.5"
    end_time = "2024-02-02 10:17:40.5"
    
    

    
    
    

plt.gca().xaxis.set_major_formatter(mdates.DateFormatter('%H:%M:%S'))
plt.xlim(pd.Timestamp(start_time), pd.Timestamp(end_time))

plt.grid()
plt.xlabel("UTC DateTime")
plt.ylabel(comp[2]+" компонента, у.е. АЦП")
plt.title('Компонента '+ comp[2])
plt.autoscale(enable=True, axis='y', tight=True)

plt.legend(loc='upper right')
plt.show()
```

```python

mask = (combined_datetime >= start_time) & (combined_datetime <= end_time)

export_data = {'Time': combined_datetime[mask].dt.strftime('%H:%M:%S.%f'), 'Ampl': m[mask]}

export_df = pd.DataFrame(export_data)


export_df.to_csv('expdatamikros.dat', sep='\t', index=False)
```

```python
for i, df in enumerate(dfs):
    combined_datetime = pd.to_datetime(df['Date_Time']) + pd.to_timedelta(df['msec'], unit='ms')
    parent_dir = os.path.basename(os.path.dirname(file_paths[i]))

    m = df[names[0]] 
    
    start_time = pd.Timestamp("2024-02-02 10:17:35.5")
    end_time = pd.Timestamp("2024-02-02 10:17:37.5")
    
    mask = (combined_datetime >= start_time) & (combined_datetime <= end_time)
    
    export_data = {'Time': combined_datetime[mask].dt.strftime('%H:%M:%S.%f')[:-3], 'Ampl': m[mask]}
    
    export_df = pd.DataFrame(export_data)
    

    export_df.to_csv(f'exported_data_{parent_dir}.dat', sep='\t', index=False)
```

```python
import os
import pandas as pd
import numpy as np
from matplotlib import pyplot as plt
import matplotlib.dates as mdates

folder_path = r"C:\Users\Alina\Desktop\3389"
file_paths = [os.path.join(folder_path, f) for f in os.listdir(folder_path) if f.endswith('.dat')]

comp = ["Z", "N", "E"]

def on_key(event):
    if event.key == 'left':
        ax_start, ax_end = axs[0].get_xlim()
        axs[0].set_xlim(ax_start - 30000, ax_end - 30000)  # Shift 5 minutes backward
        plt.draw()

    elif event.key == 'right':
        ax_start, ax_end = axs[0].get_xlim()
        axs[0].set_xlim(ax_start + 30000, ax_end + 30000)  # Shift 5 minutes forward
        plt.draw()

for file_path in file_paths:
    df = pd.read_csv(file_path, sep="\t", encoding='cp1251', skiprows=29, parse_dates=[['Date', 'Time']])
    fig, axs = plt.subplots(3, 1, figsize=(9, 10))

    for i in range(3):
        buff = df[str(i+1)]  
        axs[i].plot(df["Date_Time"], buff, 'k')
        axs[i].grid()
        axs[i].set_xlabel("UTC Datetime")
        axs[i].set_ylabel(f"{comp[i]} компонента, у.е. АЦП")
        axs[i].autoscale(enable=True, axis='both', tight=False)
        
        axs[i].xaxis.set_major_formatter(mdates.DateFormatter('%H:%M:%S'))
        axs[i].format_xdata = mdates.DateFormatter('%Y-%m-%d %H:%M:%S')
        axs[i].xaxis_date()
        
        axs[i].set_xlim([df["Date_Time"].iloc[0], df["Date_Time"].iloc[0] + pd.Timedelta(minutes=5)])
        axs[i].xaxis.set_major_locator(mdates.MinuteLocator())
        axs[i].xaxis.set_minor_locator(mdates.SecondLocator(bysecond=range(0,60,10)))
        
        fig.canvas.mpl_connect('key_press_event', on_key) 

    plt.tight_layout()
    plt.suptitle(os.path.basename(file_path))
    plt.show()
```

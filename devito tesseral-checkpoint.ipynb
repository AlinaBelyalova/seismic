```python
import numpy as np
import segyio
import matplotlib.pyplot as plt
import os

def normalize_and_plot(file1, file2):
    #    
    save_dir = r"\\server\ngen_3_\2024.07 - (2005) .  \030 - \ ( )\imgs2"
    os.makedirs(save_dir, exist_ok=True)
    
    #     
    with segyio.open(file1, "r", ignore_geometry=True) as f1:
        data1 = -segyio.tools.collect(f1.trace.raw[:])  #  
        dt1 = segyio.tools.dt(f1) * 1e-6

    with segyio.open(file2, "r", ignore_geometry=True) as f2:
        data2 = segyio.tools.collect(f2.trace.raw[:])
        dt2 = segyio.tools.dt(f2) * 1e-6

    #  
    
    norm_data1_vis = np.array([np.roll(np.nan_to_num(tr / np.max(np.abs(tr))), 3) if np.max(np.abs(tr)) > 0 else np.zeros_like(tr) for tr in data1])
    # norm_data1_vis = np.array([tr / np.max(np.abs(tr)) if np.max(np.abs(tr)) > 0 else np.zeros_like(tr) for tr in data1])
    norm_data2_vis = np.array([np.nan_to_num(tr / np.max(np.abs(tr))) if np.max(np.abs(tr)) > 0 else np.zeros_like(tr) for tr in data2])


    #    RMS 
    difference = norm_data2_vis - norm_data1_vis
    rmse = np.sqrt(np.nanmean(difference**2, axis=1))

    #        
    vmax = max(abs(norm_data1_vis).max(), abs(norm_data2_vis).max(), abs(difference).max())
    vmin = -vmax
    
    #       
    titles = ['Tesseral (RMS: {:.2f})'.format(np.mean(np.sqrt(np.nanmean(norm_data1_vis**2, axis=1)))),
              'Devito (RMS: {:.2f})'.format(np.mean(np.sqrt(np.nanmean(norm_data2_vis**2, axis=1)))),
              'Difference (RMS: {:.2f})'.format(np.mean(np.sqrt(np.nanmean((norm_data2_vis - norm_data1_vis)**2, axis=1))))]
    data_list = [norm_data1_vis, norm_data2_vis, norm_data2_vis - norm_data1_vis]
    
    fig, axes = plt.subplots(nrows=1, ncols=3, figsize=(18, 6), sharey=True)
    for idx, ax in enumerate(axes):
        # vmax = max(abs(data_list[idx]).max(), abs(data_list[idx]).min())
        # vmin = -vmax
        
        im = ax.imshow(data_list[idx].T, cmap="seismic", aspect='auto', 
                       vmin=vmin, vmax=vmax, extent=[0, data1.shape[0], data1.shape[1]*dt1, 0])
        ax.set_title(titles[idx])
        ax.set_xlabel('Trace Number')
        ax.set_ylabel('Time (s)')
    plt.colorbar(im, ax=axes.ravel().tolist(), shrink=0.95)
    overview_fig_name = os.path.join(save_dir, os.path.basename(file1).split('.')[0] + '_overview.png')
    plt.savefig(overview_fig_name)
    plt.show()

    #  RMSE
    plt.figure(figsize=(10, 6))
    plt.plot(rmse, label='RMSE per Trace', color='blue')
    plt.title("RMSE")
    plt.xlabel("Trace Number")
    plt.ylabel("RMSE")
    plt.legend()
    plt.grid(True)
    rmse_fig_name = os.path.join(save_dir, os.path.basename(file1).split('.')[0] + '_RMSE.png')
    plt.savefig(rmse_fig_name)
    plt.show()

    #    100- 
    plt.figure(figsize=(10,6))
    plt.plot(norm_data1_vis[99][:2000], lw=1.5, color='red', label='Tesseral')
    plt.plot(norm_data2_vis[99][:2000], lw=1.5, color='green', label='Devito')
    plt.legend()
    plt.title("Trace")
    plt.xlabel("Sample Number")
    plt.ylabel("Normalized Amplitude")
    plt.grid(True)
    trace_fig_name = os.path.join(save_dir, os.path.basename(file1).split('.')[0] + '_trace100.png')
    plt.savefig(trace_fig_name)
    plt.show()








# file1 = r"\\server\ngen_3_\2024.07 - (2005) .  \030 - \ ( )\2_tesseral\mod2_s0_up+GathEP-Z.sgy"

# file1 = r"\\server\ngen_3_\2024.07 - (2005) .  \030 - \ ( )\2_tesseral\mod2_s0_down+GathEP-Z.sgy"

# file1 = r"\\server\ngen_3_\2024.07 - (2005) .  \030 - \ ( )\2_tesseral\mod2_s250_100_up+GathEP-Z.sgy"

file1 = r"\\server\ngen_3_\2024.07 - (2005) .  \030 - \ ( )\2_tesseral\mod2_s250_100_down+GathEP-Z.sgy"
# file1 = r"\\server\ngen_3_\2024.07 - (2005) .  \030 - \ ( )\2_tesseral\mod2_sbound_up+GathEP-Z.sgy"
# file1 = r"\\server\ngen_3_\2024.07 - (2005) .  \030 - \ ( )\2_tesseral\mod2_sbound_down+GathEP-Z.sgy"





# file2 = r"\\server\ngen_3_\2024.07 - (2005) .  \030 - \ ( )\mod2_devito\Model 2_x250y0_top.sgy"
# file2 = r"\\server\ngen_3_\2024.07 - (2005) .  \030 - \ ( )\mod2_devito\Model 2_x250y0_bot.sgy"
# file2 = r"\\server\ngen_3_\2024.07 - (2005) .  \030 - \ ( )\mod2_devito\Model 2_x250y100_top.sgy"
file2 = r"\\server\ngen_3_\2024.07 - (2005) .  \030 - \ ( )\mod2_devito\Model 2_x250y100_bot.sgy"
# file2 = r"\\server\ngen_3_\2024.07 - (2005) .  \030 - \ ( )\mod2_devito\Model 2_x250y250_top.sgy"
# file2 = r"\\server\ngen_3_\2024.07 - (2005) .  \030 - \ ( )\mod2_devito\Model 2_x250y250_bot.sgy"

normalize_and_plot(file1, file2)
```

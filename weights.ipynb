```python
import numpy as np
from matplotlib import pyplot as plt

x = np.arange(0, 41, 1)
y = np.arange(-200, 1, 1)

xx, yy = np.meshgrid(x,y, indexing='xy')
# xx = xx *1

xx = xx-20
yy = yy+100
xx = xx *0.5


vv = (((yy - xx) * (xx + yy) / (xx ** 2 + yy ** 2 + 0.00001)) + 1) / 4
plt.imshow(vv)
plt.axvline(40)
plt.show()
```
